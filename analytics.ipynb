```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
pokemon = pd.read_csv('/content/drive/MyDrive/pokemon.csv')
pokemon.head(11)
```

```python
pokemon.describe()
```

```python
pokemon["Name"]=pokemon["Name"].str.lower()
pokemon.head(11)
```

```python
pokemon["Type 2"] = pokemon["Type 2"].fillna("Normal")
pokemon.head(11)
```

```python
pokemon["HP"].plot()
pokemon["HP"].hist()
```

```python
pokemon[ pokemon["Legendary"] == True ].head(101)
```

```python
pokemon["Legendary"] = pokemon.loc[pokemon["Type 1"] == "Fire", "Type 1"].head(11)
print(pokemon["Legendary"]) # Changed from print["Legendary"] to print(pokemon["Legendary"])
#  .inplace(True) is not needed as assignment creates a new column
```

```python
import matplotlib.pyplot as plt

# Assuming 'pokemon' is your DataFrame
total_pokemon = len(pokemon)

# Correctly filter for Legendary Pokemon (assuming 'Legendary' column is boolean)
num_legendary = len(pokemon[pokemon["Legendary"] == True])
num_non_legendary = total_pokemon - num_legendary

labels = ['Legendary', 'Non-Legendary']
sizes = [num_legendary, num_non_legendary]

plt.figure(figsize=(4, 4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Legendary vs. Non-Legendary Pokemon')
plt.axis('equal')  # Ensures the pie chart is circular
plt.show()
```

```python
legendary_pokemon = pokemon[pokemon["Legendary"] == True]
plt.figure(figsize=(12, 6))
sns.boxplot(data=legendary_pokemon[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']])
plt.title('Distribution of Stats for Legendary Pokemon')
plt.xlabel('Stats')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(legendary_pokemon['Total'], bins=20)
plt.title('Distribution of Total Stats for Legendary Pokemon')
plt.xlabel('Total Stats')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Attack', y='Defense', data=legendary_pokemon, hue='Type 1')
plt.title('Attack vs. Defense for Legendary Pokemon')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.show()
```
